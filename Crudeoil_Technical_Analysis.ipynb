```
import quandl
import matplotlib.pyplot as plt
%matplotlib inline

data = quandl.get('EIA/PET_RWTC_D',start_date='2017-1-1',end_date='2020-1-1')
data['Value'].plot(figsize=(40,40))

# plotting 20 days moving average
data.rolling(window=20).mean()['Value'].plot()

# plotting Donchian Channel Indicator
data.rolling(window=20).max()['Value'].plot()
data.rolling(window=20).min()['Value'].plot()
```
